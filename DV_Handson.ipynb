```python
# Import the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the dataset
df_student = pd.read_csv(r'D:\Intellipaat\python resources\student.csv')
df_student
```

### Edit Dataframe

```python
# Renaming 'race/ethnicity' column to 'ethnicity'
df_student.rename(columns={'race/ethnicity' : 'ethnicity'}, inplace=True)

# Renaming 'parental level of education' to 'parent's education'
df_student.rename(columns={'parental level of education' : "parent's education"}, inplace=True)

# Rename 'test preparation course' to 'preparation test'
df_student.rename(columns={'test preparation course' : 'preparation test'}, inplace=True)
```

```python
# Creating 'total marks' column having sum of all scores
df_student['total score'] = df_student['math score'] + df_student['reading score'] + df_student['writing score']

# Creating 'percentage' column having mean of all scores
df_student['percentage'] = (df_student['total score']/3).round(2)
```

### Ploting

```python
# Count sub plots

# Grid
sns.set_style(style='darkgrid')

# No. of plots & Dimensions
fi, axs = plt.subplots(1, 2, figsize=(20, 7))

# Plot
sns.countplot(x='gender', data=df_student, ax=axs[0], hue='gender', palette='Dark2')
sns.countplot(x="ethnicity", data=df_student, ax=axs[1], hue='ethnicity', palette='magma', 
              order=['group A', 'group B', 'group C', 'group D', 'group E'])

# Show
plt.show()
```

```python
# Histogram of math score

# Grid
sns.set_style(style='darkgrid')

# Plot
sns.histplot(df_student['math score'], bins=30, color='red', kde=True)

# Show
plt.show()

# By ploting the graph shows that it is normally distributed.
```

```python
# Joint plot for reading score & writing score

# Plot
sns.jointplot(x='reading score', y='writing score', data=df_student, kind='reg')

# Show
plt.show()
```

```python
# Pie plot for ethnicity

# Data
ethnicity_count = df_student['ethnicity'].value_counts()

# Plot
plt.pie(ethnicity_count, labels=ethnicity_count.index, autopct='%1.1f%%', radius=1.2)

# Show
plt.show()
```

### Importing Dataset

```python
df_cars = pd.read_csv('cars.csv')
df_cars
```

### Analyse Dataframe (before editing)

```python
df_cars.info()
```

### Editing Dataframe

```python
df_cars['cubicinches'] = pd.to_numeric(df_cars['cubicinches'], errors='coerce')
df_cars['weightlbs'] = pd.to_numeric(df_cars['weightlbs'], errors='coerce')
```

### Analyse Dataframe (after editing)

```python
df_cars.info()
```

### Ploting

```python
# Create scatter plot for the following data, make sure all the plot apear in the same plane with x-label & y-label evenly spaced
# 1.. 'hp' & 'mpg'
# 2.. 'hp' & 'time-to-60'
# 3.. 'hp' & 'weightlbs'
# 4.. 'hp' & 'cylinders'

fig, axes = plt.subplots(2, 2, figsize = (20, 15))

# 1. 'hp' & 'mpg'
plt.subplot(2, 2, 1)
sns.scatterplot(x='hp', y='mpg', data=df_cars)
plt.xlabel('horse power')
plt.ylabel('miles per gallon')
plt.title('hp & mpg')

# 2 'hp' & 'time-to-60'
plt.subplot(2, 2, 2)
sns.scatterplot(x='hp', y='time-to-60', data=df_cars)
plt.xlabel('horse power')
plt.ylabel('time-to-60')
plt.title('hp & time-to-60')

# 3 'hp' & 'weightlbs'
plt.subplot(2, 2, 3)
sns.scatterplot(x='hp', y='weightlbs', data=df_cars)
plt.xlabel('horse power')
plt.ylabel('weightlbs')
plt.title('hp & weightlbs')

# 4 'hp' & 'cylinders'
plt.subplot(2, 2, 4)
sns.scatterplot(x='hp', y='cylinders', data=df_cars)
plt.xlabel('horse power')
plt.ylabel('cylinders')
plt.title('hp & cylinders')

plt.show()
```

```python
# Create scatter plot for following data with respect to each brand.
# 1.. 'mpg' & 'hp'
# 2.. 'mpg' & 'weight'
# 3.. 'hp' & 'mpg'
# 4.. 'hp' & 'time-to-60'

fig, axes = plt.subplots(2, 2, figsize = (20, 15))

# 1. 'mpg' & 'hp'
plt.subplot(2, 2, 1)
sns.scatterplot(x='mpg', y='hp', data=df_cars, hue='brand')
plt.title('mpg & hp')

# 2. 'mpg' & 'weight'
plt.subplot(2, 2, 2)
sns.scatterplot(x='mpg', y='weightlbs', data=df_cars, hue='brand')
plt.title('mpg & weight')

# 3. 'hp' & 'mpg'
plt.subplot(2, 2, 3)
sns.scatterplot(x='hp', y='mpg', data=df_cars, hue='brand')
plt.title('hp & mpg')

# 4. 'hp' & 'time-to-60'
plt.subplot(2, 2, 4)
sns.scatterplot(x='hp', y='time-to-60', data=df_cars, hue='brand')
plt.title('hp & time-to-60')

plt.show()
```

```python
# Create bar plot for the following, make sure all the plots appear in same plane with x-axis & y-axis evenly spaced.
# 1.. cylinder & horsepower, cylinders & miles per gallon, cylinders and weight.
# 2.. year & time-to-60, year & miles per gallon, year & horsepower.

fig, axes = plt.subplots(3, 2, figsize = (20, 20))

# 1. cylinder & horsepower
plt.subplot(3, 2, 1)
sns.barplot(x='cylinders', y='hp', data=df_cars)
plt.title('Cylinders & Horsepower')
plt.xlabel('Cylinders')
plt.ylabel('Horsepower')

# 2. cylinders & miles per gallon
plt.subplot(3, 2, 2)
sns.barplot(x='cylinders', y='mpg', data=df_cars)
plt.title('Cylinders & Miles per gallon')
plt.xlabel('Cylinders')
plt.ylabel('Miles per gallon')

# 3. cylinders and weightlbs
plt.subplot(3, 2, 3)
sns.barplot(x='cylinders', y='weightlbs', data=df_cars)
plt.title('Cylinders & Weightlbs')
plt.xlabel('Cylinders')
plt.ylabel('Weightlbs')

# 4. year & time-to-60
plt.subplot(3, 2, 4)
sns.barplot(x='year', y='time-to-60', data=df_cars)
plt.title('Year & Time-to-60')
plt.xlabel('Year')
plt.ylabel('Time-to-60')

# 5. year & miles per gallon
plt.subplot(3, 2, 5)
sns.barplot(x='year', y='mpg', data=df_cars)
plt.title('Year & Miles per gallon')
plt.xlabel('Year')
plt.ylabel('Miles per gallon')

# 6. year & horsepower
plt.subplot(3, 2, 6)
sns.barplot(x='year', y='hp', data=df_cars)
plt.title('Year & Horsepower')
plt.xlabel('Year')
plt.ylabel('Horsepower')

plt.show()
```
