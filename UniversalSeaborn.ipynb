# **Welcome to Seaborn**

```python
# Import the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Simple line plot

```python
# Sample data
x = [1, 2, 3, 4, 5, 6, 7]
y = [4, 5, 1, 7, 6, 8, 9]

# Plot 
sns.lineplot(x=x, y=y)

# Show
plt.show()
```

### Load dataset

```python
# Import the required dataset
data = pd.read_csv(r'D:\Intellipaat\python resources\student.csv').head(25)
data
```

### Line plot demo

```python
# Plot
sns.lineplot(x='reading score', y='writing score', data=data, hue='gender', palette='magma', 
            style='gender' ,markers=['o', 'd'])

# Grid
plt.grid(True, linestyle = '--')

# Add label
plt.title("Line plot demo")

# Show 
plt.show()
```

### Histogram plot

```python
# Plot
sns.histplot(data['math score'], color='skyblue')

# Show
plt.show()
```

### Bar plot

```python
# Plot
ord = ['group A', 'group B', 'group C', 'group D']
sns.barplot(x='race/ethnicity', y='writing score', data=data, hue='gender', saturation=100, 
            order=ord, err_kws={'linewidth' : 0})

# Add label
plt.title("Bar plot")

# Show
plt.show()
```

### Scatter plot

```python
# Dimension
plt.figure(figsize=(9, 6))

# Plot
sns.scatterplot(x='reading score', y='writing score', data=data, hue='gender', style='gender', 
                size='gender', sizes=(60, 40), palette='ocean')

# Add labels
plt.title("Scatter plot")

# Show
plt.show()
```

### Heatmap

```python
# Sample data
arr = np.linspace(1, 10, 25).reshape(5, 5)

# Plot
sns.heatmap(arr, cmap='Purples', annot=True, linecolor='black', linewidths=0.4)


# Add labels
plt.title("Heatmap")

# Show 
plt.show()
```

### Countplot

```python
# Plot
ord = ['group A', 'group B', 'group C', 'group D']
sns.countplot(x='gender', data=data, hue='race/ethnicity', hue_order=ord)

# Add labels
plt.title("Countplot")

# Show 
plt.show()
```

### Violin plot

```python
# Plot
ord = ['group A', 'group B', 'group C', 'group D']
sns.violinplot(x='race/ethnicity', y='reading score', data=data, hue='gender', palette='magma',
               order=ord, inner='stick')

# Add labels
plt.title("Violin plot")

# Show
plt.show()
```

### Pair plot

```python
# Plot 
sns.pairplot(data, hue='gender')

# Show 
plt.show()
```

### Strip plot

```python
# Plot
sns.stripplot(x='test preparation course', y='math score', data=data, hue='gender', 
              palette='Set1', edgecolor='black', linewidth=1)

# Add label
plt.title("Strip plot")

# Show
plt.show()
```

### Box plot

```python
# Dimension & Fonts
sns.set_theme(font_scale=1.4, rc={'figure.figsize' : (15, 8)})

# Plot
sns.boxplot(x='parental level of education', y='math score', data=data, hue='parental level of education', 
            palette='magma')

# Show
plt.show()
```

### Plot 'point plot' using cat plot

```python
# Plot
sns.catplot(x='lunch', y='reading score', data=data, hue='gender',height=5, aspect=2, kind='point', palette='magma')

# Add label
plt.title("Cat Plot")

# Show
plt.show()
```

### Plot 'swarm plot' using cat plot

```python
# Plot
sns.catplot(x='lunch', y='reading score', data=data, hue='gender',height=5, aspect=2, kind='swarm', palette='magma')

# Add label
plt.title("Cat Plot")

# Show
plt.show()
```

### Seaborn plot Demo (style & customize)

```python
# Set grid
sns.set_style('whitegrid')

# Dimension
plt.figure(figsize=(10, 5))

# Fonts
sns.set_context('paper', font_scale=1.5)

# Plot
sns.barplot(x='race/ethnicity', y='math score', data=data, hue='gender', palette='ocean', 
            order=['group A', 'group B', 'group C', 'group D'])

# Add label
plt.title("Seaborn Plot Demo")

# Show
plt.show()
```

### Subplots (customized axs)

```python
# Grid
sns.set_style('whitegrid')

# Subplots & Figsize
fi, axs = plt.subplots(2, 2, figsize=(30, 15))      # Here first 2 is for rows & second 2 is for columns for graph

# Plot
sns.barplot(x='gender', y='math score', data=data, hue='gender', palette='Dark2', 
            ax=axs[0, 0])                                                       # 0th row & 0th column
sns.barplot(x='test preparation course', y='math score', data=data, hue='gender', 
            palette='ocean', ax=axs[0, 1])                                      # 0th row & 1st column
sns.barplot(x='parental level of education', y='math score', data=data, hue='gender', 
            palette='magma', ax=axs[1, 0])                                      # 1st row & 0th column
sns.barplot(x='race/ethnicity', y='math score', data=data, hue='gender', palette='Accent', 
            order=['group A', 'group B', 'group C', 'group D'], ax=axs[1, 1])   # 1st row & 1st column

# Show
plt.show()
```

### Multiple plot (FacetGrid function)

```python
# 
sns.set_context('paper', font_scale=1.1)

# Plot
fg = sns.FacetGrid(data, col='lunch', row='parental level of education', sharey=False,
                   row_order=['high school', 'some college', "associate's degree", "bachelor's degree", 
                              "master's degree"])

fg.map_dataframe(sns.barplot, x = 'gender', y = 'math score', color = 'skyblue')

fg.set_axis_labels('Gender', 'Math Score')

fg.set_titles(col_template='{col_name} lunch', row_template='{row_name}')
```

### Joint plot

```python
# Plot
sns.jointplot(x='reading score', y='writing score', data=data, color='red', height=6, kind='reg')

# Show
plt.show()
```
