# **Welcome to Matplotlib**

```python
# Import the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

### Simple Line plot

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 6]

# Plot graph
plt.plot(x, y)

# Add labels
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.title("Simple Line Plot")

# Grid
plt.grid(True)

# Show
plt.show()
```

### Line Plot Demo

```python
# Sample data
x = np.arange(0, 10, 1)
y = 2 * x + 20

# Dimension of plot
fig = plt.figure(figsize=(10, 5))

# Plot graph
plt.plot(x, y)

# Add labels
plt.xlabel(" X Data")
plt.ylabel("Y Data")
plt.title("Line PLot Demo")
```

### Fancy Line Plot 

```python
# Sample data
x = np.arange(0, 10, 1)
y = 2 * x + 20

# Dimension of plot
fig = plt.figure(figsize=(10, 5))                   # figsize = (width, height)

# Plot graph
plt.plot(x, y, linestyle = ':', color = 'red', alpha = 0.5, marker = 'o')

# Add labels
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.title("Fancy line plot ")

# Legend
plt.legend(["Growth Line"], loc = 'best')

# Grid
plt.grid(True)

# Show
plt.show
```

### Multple Line Plot

```python
# Sample data
x = np.arange(0, 10, 1)
y1 = 2 * x + 5
y2 = 2 * x + 10

# Dimension of plot
plt.figure(figsize=(10, 5))

# Plot graph
    # plot y1
plt.plot(x, y1, linestyle = ':', color = 'red', alpha = 0.5, marker = 'o', label = 'Lane 1')

    # plot y2
plt.plot(x, y2, linestyle = '-', color = 'blue', alpha = 0.5, marker = 'd', label = 'Lane 2')

# Add labels
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Multiple Line Plot')

plt.legend()

# Grid
plt.grid(True)

# Show
plt.show()
```

### Multiple Area Plot

```python
# Sample data
x = [1, 2, 3 , 4, 5]
y1 = [10, 12, 4, 5, 6]
y2 = [8, 6, 5, 4, 14]

# Dimension of plot
plt.figure(figsize=(8, 6))

# Plot graph
    # Customization y1 
plt.fill_between(x, y1, color = 'red', alpha = 0.5, label = 'Poogle Pvt Ltd')

    # Customization y2
plt.fill_between(x, y2, color = 'green', alpha = 0.5, label = 'Touch Inc')

# Labels
plt.title("Multiple Area Plot")

# Legend
plt.legend(loc = 'upper left')

# Grid
plt.grid(True)

# Show
plt.show()
```

### Simple Bar Plot

```python
# Sample data
months = ['jan', 'feb', 'mar', 'apr', 'may', ]
product_sold = [100, 90, 78, 57, 65]

# Plot graph
plt.bar(months, product_sold, color = 'skyblue')

# Add labels
plt.xlabel('Months')
plt.ylabel("Product sold")
plt.title("Simple bar plot")

# Show
plt.show()
```

### Horizontal Bar Plot

```python
# Sample data
data = {"Sofa" : 40, 
        "Bed" : 30, 
        "Desk" : 50, }

names = list(data.keys())
quantity = list(data.values())

# Dimension of plot
plt.figure(figsize=(8, 4))

# Plot graph
plt.barh(names, quantity, color = 'lightblue')

# Add labels
plt.xlabel('Quantity')
plt.ylabel('Products')
plt.title("Horizontal Bar Plot")

# Show
plt.show()
```

### Simple Scatter Plot

```python
# Sample data
x = [10, 20, 30, 40, 50]
y = [1, 2.5, 3, 4.2, 3.7]

# Plot graph
plt.scatter(x, y)

# Show
plt.show()
```

### Scatter Plot Demo

```python
# Sample data
np.random.seed(0)               # Generates random numbers
x = np.random.random(50)        # Random 50 numbers between 0 & 1
y = np.random.random(50)        # -------------------------------

# Dimension of plot
plt.figure(figsize=(8, 6))

# Plot
plt.scatter(x, y, marker='o', color = 'red', edgecolors='blue', linewidths=1.5, alpha=0.7, label = 'Data Point')

# Add labels
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title("Scatter Plot Demo")

# Legend
plt.legend()

# Grid
plt.grid(True, linestyle = '--', alpha = 0.5)

# Limits of axis
plt.xlim(0, 1)                  # Limit of X Axis
plt.ylim(0, 1)                  # Limit of Y Axis

# Show
plt.show()
```

### Histogram Plot

```python
# Sample data
np.random.seed(8)
data = np.random.random(100)

# Plot graph
plt.hist(data, bins=25, color='skyblue', edgecolor='black')     # 'HEX' color codes can also be used to change 
                                                                #  the color of bars.
# Add labels
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title("Histogram Plot")

# Grid
plt.grid(True, linestyle = '--', alpha = 0.5)

# Show
plt.show()
```

### Box Plot

```python
# Sample data
section_A = [81, 88, 79, 77, 91, 94, 98, 97, 99, 89, 84, 85]
section_B = [61, 66, 64, 73, 75, 71, 69, 69, 67, 56, 68, 71]
section_C = [44, 48, 56, 60, 59, 57, 52, 42, 39, 48, 49, 55]
data = list([section_A, section_B, section_C])

# Plot graph
plt.boxplot(data, showmeans=True)

# Add labels
plt.xlabel('X-Axis')
plt.ylabel('Percentage')
plt.title("Box Plot")

# Grid
plt.grid(True)

# Show                  # green mini triangle is 'mean'
plt.show()              # Boxes plot shows distribution of data in 'box' & also shows outliers in 'circle'
```

### Violin plot

```python
# Sample data
section_A = [81, 88, 79, 77, 87, 94, 98, 97, 99, 89, 84, 85]
section_B = [61, 66, 64, 73, 75, 71, 69, 69, 67, 56, 68, 71]
section_C = [44, 48, 56, 60, 59, 57, 52, 42, 39, 48, 49, 55]
data = list([section_A, section_B, section_C])

# Plot graph
plt.violinplot(data, showmeans=True)

# Add labels
plt.xlabel('X-Axis')
plt.ylabel('Percentage')
plt.title("Violin Plot")

# Grid
plt.grid(True)

# Show
plt.show()              # Violin plot show distribution of data in 'density curves'
```

### Simple pie chart / plot

```python
# Sample data
titles = [5, 5, 2, 1, 1, 1, 1]
teams = ['CSK', 'MI', 'KKR','RR', 'SRH', 'GT', 'DC']
color = ['yellow', 'blue', 'purple', 'pink', 'orange', 'black', 'skyblue']

# Plot graph 
plt.pie(titles, labels=teams, colors=color, radius=1.2)

# Add labels
plt.title("Simple pie plot")

# Show
plt.show()
```

### Pie chart / plot Demo

```python
# Sample data
runs = [973, 840, 821, 760, 720]
players = ['V.Kohli', 'S.Gill', 'J.Buttler', 'YB.Jaiswal', 'S.Dhawan']
total_runs = sum(runs)
exp = [0.1, 0, 0, 0, 0]

# Func to add 'runs' inside pie chart
def autopct_runs(prct_value):                               # prct_value => percentage value get by 'autopct' func
    return(f"{(total_runs * prct_value) / 100:.0f}")

# Plot graph
plt.pie(runs, labels=players, autopct=autopct_runs, shadow=True, explode=exp, radius=1.2, 
        textprops={'fontsize' : 12}, wedgeprops={'linewidth' : 1, 'edgecolor' : 'black'})

# Add labels
plt.title("Pie Plot Demo")

# Legend
# plt.legend(loc = 'lower right')

# Show
plt.show()
```
